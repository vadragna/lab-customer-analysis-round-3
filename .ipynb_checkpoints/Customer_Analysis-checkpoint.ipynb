```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
display(data.head())
data.describe().T
data.describe(include=[object]).T

response_counts = data['Response'].value_counts()
print("response_counts", response_counts)

unique_responses = response_counts.index.tolist()
counts = response_counts.tolist()

plt.bar(unique_responses, counts)

plt.xlabel("Response")
plt.ylabel("Total Number of Responses")
plt.title("Total Number of Responses for Each Category")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

data["Sales Channel"].unique()

filtered_agent = data[data['Sales Channel'] == 'Agent']
filtered_call_center = data[data['Sales Channel'] == 'Call Center']
filtered_call_web = data[data['Sales Channel'] == 'Web']
filtered_call_branch = data[data['Sales Channel'] == 'Branch']

response_counts_agent = filtered_agent['Response'].value_counts(normalize=True) * 100
response_call_center = filtered_call_center['Response'].value_counts(normalize=True) * 100
response_call_web = filtered_call_web['Response'].value_counts(normalize=True) * 100
response_branch = filtered_call_branch['Response'].value_counts(normalize=True) * 100

print("response_counts agent", response_counts_agent, response_call_center, response_call_web, response_branch)

response_rates = {
    'Agent': {'No': 80.845557, 'Yes': 19.154443},
    'Call Center': {'No': 88.546942, 'Yes': 11.453058},
    'Web': {'No': 88.226415, 'Yes': 11.773585},
    'Branch': {'No': 88.546942, 'Yes': 11.453058}
}

# Convert the response rates to a DataFrame for easier plotting
response_rates_df = pd.DataFrame(response_rates)

# Plot only the 'Yes' response rates using a bar plot
response_rates_df.loc['Yes'].plot(kind='bar')

# Set plot labels and title
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate (%)')
plt.title('Response Rate (Yes) by Sales Channel')

# Show the plot
plt.show()




```

```python
bins = [0, 300, 600, 900, 1200, float('inf')]
labels = ['0 - 300', '301 - 600', '601 - 900', '901 - 1200', 'From 1201']

data['Claim Amount Bins'] = pd.cut(data['Total Claim Amount'], bins=bins, labels=labels)

sns.countplot(x='Claim Amount Bins', data=data, order=labels)
plt.xlabel('Total Claim Amount Bins')
plt.ylabel('Count')
plt.title('Distribution of Total Claim Amount')
plt.show()

dfs_by_bin = {}

for bin_label in labels:
    dfs_by_bin[bin_label] = data.loc[data['Claim Amount Bins'] == bin_label]

# Ora puoi accedere ai DataFrame filtrati per ogni bin usando il dizionario
dfs_by_bin['0 - 300'] = filtered_agent['Response'].value_counts(normalize=True) * 100
dfs_by_bin['301 - 600'] = filtered_call_center['Response'].value_counts(normalize=True) * 100
dfs_by_bin['601 - 900'] = filtered_call_web['Response'].value_counts(normalize=True) * 100
dfs_by_bin['901 - 1200'] = filtered_call_branch['Response'].value_counts(normalize=True) * 100
dfs_by_bin['From 1201'] = filtered_call_branch['Response'].value_counts(normalize=True) * 100

print(dfs_by_bin['0 - 300'])

print(dfs_by_bin['0 - 300'].head)
print(dfs_by_bin['301 - 600'].head)
print(dfs_by_bin['601 - 900'].head)
print(dfs_by_bin['901 - 1200'].head)
print(dfs_by_bin['From 1201'].head)

response_rates = {
    '0 - 300': {'No': 80.845557, 'Yes': 19.154443},
    '301 - 600': {'No': 89.121813, 'Yes': 10.878187},
    '601 - 900': {'No': 88.226415, 'Yes': 11.773585},
    '901 - 1200': {'No': 88.546942, 'Yes': 11.453058},
    'From 1201': {'No': 88.546942, 'Yes': 11.453058}
}

response_rates_df = pd.DataFrame(response_rates)

response_rates_df.loc['Yes'].plot(kind='bar')

plt.xlabel('Total Claim Account Bin')
plt.ylabel('Response Rate (%)')
plt.title('Response Rate (Yes) by Total Claim Account Bin')

# Show the plot
plt.show()

```

```python
response_rates_TCA = data[data['Response'] == 'Yes']['Income'].value_counts() / data['Income'].value_counts()

plt.figure(figsize=(10, 6))
sns.histplot(response_rates_TCA, kde=True, label='Yes')
plt.title('Distribution of Income by Response')
plt.xlabel('Income')
plt.ylabel('Density')
plt.show()
```

```python
income_bins = [0, 30000, 50000, 70000, 100000, float('inf')]
income_labels = ['0 - 30000', '30001 - 50000', '50001 - 70000', '70001 - 100000', 'Above 100000']

data['Income Bins'] = pd.cut(data['Income'], bins=income_bins, labels=income_labels, right=False)

dfs_by_income_bin = {}

for income_label in income_labels:
    dfs_by_income_bin[income_label] = data.loc[data['Income Bins'] == income_label]
    
dfs_by_income_bin['0 - 30000'] = dfs_by_income_bin['0 - 30000']['Response'].value_counts(normalize=True) * 100
dfs_by_income_bin['30001 - 50000'] = dfs_by_income_bin['30001 - 50000']['Response'].value_counts(normalize=True) * 100
dfs_by_income_bin['50001 - 70000'] = dfs_by_income_bin['50001 - 70000']['Response'].value_counts(normalize=True) * 100
dfs_by_income_bin['70001 - 100000'] = dfs_by_income_bin['70001 - 100000']['Response'].value_counts(normalize=True) * 100
dfs_by_income_bin['Above 100000'] = dfs_by_income_bin['Above 100000']['Response'].value_counts(normalize=True) * 100

print(dfs_by_income_bin['0 - 30000'])
print(dfs_by_income_bin['30001 - 50000'])
print(dfs_by_income_bin['50001 - 70000'])
print(dfs_by_income_bin['70001 - 100000'])
print(dfs_by_income_bin['Above 100000'])

response_rate = {
    '0 - 30000': {'No': 80.845557, 'Yes': 19.154443},
    '30001 - 50000': {'No': 89.121813, 'Yes': 10.878187},
    '50001 - 70000': {'No': 88.226415, 'Yes': 11.773585},
    '70001 - 100000': {'No': 88.546942, 'Yes': 11.453058},
    'Above 100000': {'No': 88.546942, 'Yes': 11.453058}
}

response_rates_df = pd.DataFrame(response_rates)
response_rates_df.loc['Yes'].plot(kind='bar')

plt.xlabel('Total Income Bin')
plt.ylabel('Response Rate (%)')
plt.title('Response Rate (Yes) by Total Income Bin')

plt.show()


```
